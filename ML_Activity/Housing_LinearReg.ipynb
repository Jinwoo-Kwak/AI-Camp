```python
def cost(x, y, w, b): 
    m = x.shape[0]
    cost = 0.0
    for i in range(m):                                
        prod = np.dot(x[i], w) + b           
        cost = cost + (prod - y[i])**2
    cost = cost / (2 * m)                          
    return cost
```

```python
def gradient(X, y, w, b): 
    m = X.shape[0]
    n = X.shape[1] # weight for dot product      
    grad_w = np.zeros((n,)) #gradient for w
    grad_b = 0 # gradient for b

    for i in range(m):                             
        loss = (np.dot(X[i], w) + b) - y[i]   
        for j in range(n):                         
            grad_w[j] = grad_w[j] + loss * X[i, j]    
        grad_b = grad_b + loss                        
    grad_w = grad_w / m                                
    grad_b = grad_b / m                                
        
    return grad_b, grad_w
```

```python
def gradient_descent(X, y, w, b, alpha,num_iters, cost_function, gradient_function):  
    for i in range(num_iters):
        dj_db,dj_dw = gradient_function(X, y, w, b)   

        w = w - alpha * dj_dw               
        b = b - alpha * dj_db               
          
        if i% 100 == 0:
            print('Iteration: ' + str(i) + ', Loss: ' + str(cost_function(X,y,w,b)))
        
    return w, b
```
