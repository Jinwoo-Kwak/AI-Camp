```python
# add rows & columns

# two dataframe
#  - append, extend, concatencate

df1 = pd.DataFrame({'A' : range(4),
                    'B': range(4,8),
                    'C' : range(9,13),
                    'D' : range(13,17)})

df1
```

```python
df2 = pd.DataFrame({'A' : range(40, 44),
                    'B': range(44,48),
                    'C' : range(49,53),
                    'D' : range(53,57)})

df2
```

```python
print(df1)
print(df2)
```

```python
df3 = pd.concat([df1, df2])
df3

print(df3)
df3.loc[2]
```

```python
df4 = pd.concat([df1, df2], axis=1)
print(df4)
```
